```python
def missing_value_summary(df):
    # Calculate the number of missing values for each column
    num_missing = df.isnull().sum()

    # Calculate the percentage of missing values for each column
    pct_missing = (num_missing / len(df)) * 100

    # Create a new DataFrame to store the summary
    summary_df = pd.DataFrame({'missing': num_missing, 'percent_missing': pct_missing})

    # Format the 'num_rows_missing' column as a whole percentage
    summary_df['percent_missing'] = summary_df['percent_missing'].apply(lambda x: f'{x:.0f}%')

    return summary_df.sort_values('missing', ascending= False)

# Assuming you have a DataFrame called 'df' containing your data
summary = missing_value_summary(df)

# Print the summary DataFrame
print(summary)
```

```python
import pandas as pd

def compare_columns(df, col_a, col_b):
    # Count the number of matching values
    matching_values = (df[col_a] == df[col_b]).sum()

    # Count the number of non-matching values
    non_matching_values = (df[col_a] != df[col_b]).sum()

    # Count the number of null values in col_a
    null_count_col_a = df[col_a].isnull().sum()

    # Count the number of null values in col_b
    null_count_col_b = df[col_b].isnull().sum()

    print(f"Number of matching values between {col_a} and {col_b}: {matching_values}")
    print(f"Number of non-matching values between {col_a} and {col_b}: {non_matching_values}")
    print(f"Number of null values in {col_a}: {null_count_col_a}")
    print(f"Number of null values in {col_b}: {null_count_col_b}")


# Example usage:
# Assuming you have already loaded your DataFrame df
compare_columns(df, 'calculatedbathnbr', 'bathroomcnt')

```

```python
def fips_map(df):
    """
    Map FIPS codes to county and state names and add them as new columns to a DataFrame.

    Parameters:
        df (pd.DataFrame): The DataFrame containing Zillow property data to be processed.

    Returns:
        pd.DataFrame: The DataFrame with two new columns 'county' and 'state' added based on the 'fips' column.
    """
    # Mapping of FIPS codes to county and state names
    fips_to_county_state = {
        6059: {'county': 'Orange', 'state': 'CA'},
        6111: {'county': 'Ventura', 'state': 'CA'},
        6037: {'county': 'Los Angeles', 'state': 'CA'},
    }

    # Add new columns 'county' and 'state' based on 'fips'
    df['county'] = df['fips'].map(lambda x: fips_to_county_state[x]['county'])
    df['state'] = df['fips'].map(lambda x: fips_to_county_state[x]['state'])

    return df
```
